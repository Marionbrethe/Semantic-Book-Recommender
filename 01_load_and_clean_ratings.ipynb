```python
import pandas as pd
import numpy as np
from pathlib import Path

# Set paths relative to the project root
PROJECT_ROOT = Path(".").resolve()
DATA_DIR = PROJECT_ROOT / "data"
CSV_PATH = DATA_DIR / "James_rated_books.csv"

CSV_PATH
```

```python
# List of books I've read and rated 
df_raw = pd.read_csv(CSV_PATH, sep=None, engine="python")
df_raw.head()
```

```python
df_raw.columns.tolist()
    
```

```python
cols_to_keep = [
    "Book Id",
    "Title",
    "Author",
    "ISBN",
    "ISBN13",
    "My Rating",
    "Date Read",
    "Date Added",
    "My Review"
]

df = df_raw[cols_to_keep].copy()
df.head()
```

```python
df = df.rename(columns={
    "Book Id": "book_id",
    "Title": "title",
    "Author": "author",
    "ISBN": "isbn",
    "ISBN13": "isbn13",
    "My Rating": "my_rating",
    "Date Read": "date_read",
    "Date Added": "date_added",
    "My Review": "my_review"
})

# Convert ratings to numeric
df["my_rating"] = pd.to_numeric(df["my_rating"], errors="coerce") 

# Keep only books you actually rated (my_rating > 0)
df = df[df["my_rating"] > 0].reset_index(drop=True)

df.head()
```

```python
print("Number of rated books:", len(df))
print(df[["title", "author", "my_rating"]].head(10))
```

```python
CLEAN_CSV_PATH = DATA_DIR / "my_rated_books_clean.csv"
df.to_csv(CLEAN_CSV_PATH, index=False)
CLEAN_CSV_PATH
```
